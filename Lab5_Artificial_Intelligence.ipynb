# What is social choice?

* Provides methods to make a **choice** for a
society, based on the **preferences** of the
individuals over a set of candidates

* Many kind of choices
 * **One** of the candidates
 * A **set** of candidates
 * A **ranking** over the candidates

## Voting Systems

* Set of **candidates**
* Profile: **collection** of individual’s **preferences** over the candidates
* A voting rule takes a profile and returns a **winning candidate** or
a subset of candidates
* It defines
 * The ballot structure (how to express the individuals’
preferences)
 * How to aggregate the preferences to get the result

# Which is best?

* Social choice theory looks at the (desirable and undesirable)
properties they possess
* For instance, is the rule “monotonic”?
* Bottom line: with more than 2 candidates, there is no best
voting rule for all possible cases

# Axiomatic approach

* Some desirable properties of voting rule:
 * **Anonymity**: names of voters irrelevant
 * **Neutrality**: name of candidates irrelevant
 * **Monotonicity**: if a particular candidate wins, and a voter
improves their vote in favour of this candidate, then they still win
 * **Unanimity**: If everyone votes for one candidate, they win
 * **Independent to irrelevant alternatives** (IIA): result between A and
B only depends on how agents preferences between A and B
 * **Non-dictatorial**: a dictator is voter whose vote is the result
 * **Non-imposition**: every result is possible

# Arrow’s impossibility theorem
**Theorem**: If there are at least two voters and three or more candidates, then it is impossible for any voting rule to enjoy the following properties:
* Unanimity
* Non-dictatoriality
* Independency to irrelevant alternatives

## Formal definition + examples

Let $A$ be a set of **outcomes**, $N$ a number of **voters**. $L(A)$ is the set of all full linear orderings. 

A tuple $R = (R_1, ..., R_N) \in L(A) ^ N$ is a preference profile.

We use $a \ge_{R_i} b$ to specify **preferences**, i.e., according to the rankings of the i-th voter, they prefer a over b.

E.g.: 
* There are three candidates, $A = \{a,b,c\}$
* There are two voters, $N=2$
* The set of all possible rankings of $a,b,c$ has 6 elements: $L(A) = \{(a,b,c), (a,c,b), (b,a,c), (b,c,a), (c,a,b), (c,b,a)\}$
* The first voter likes $a$ over $b$ over $c$: $R_1 = (a,b,c)$.
* The second voter $R_2 = (c, a, b)$ likes $c$ over $a$ over $b$. 
* $(R_1, R_2) \in L(A)^N$ form a preference profile of all the voters.

A preference aggregation rule is the following function:

$f : L(A)^N \rightarrow L(A) $

E.g., with $f(R) = \text{borda voting} $, then, with the example given,
$f(R) = (a,c,b)$ (since $a$ gets 3 pts, $b$ gets 1 pt, $c$ gets 2 pts).

### Unanimity

If, ($\forall R$), $\forall i \in \{1,..,N\}, a \ge_{R_i} b$, then $a \ge_{f(R)} b$

E.g.: $f(R) = R_1$ is unanimous, since, if everybody accepts that $a$ is better than $b$, then, also $R_1$ accepts it.

### Non-dictatorship
No individual $i$ whose preferences always prevail exists.

$\nexists i \in \{1, ..., N\}$ such that, $\forall a, b \in A$ if $a \ge_{R_i} b$, then $ a \ge_{f(R)} b$

E.g.: $f(R) = R_1$ is dictatorial, since $R_1$ vote always prevail.

### IIA
For two preference profiles, $R$ and $R'$, if $a \ge_{R_i} b \Leftrightarrow a \ge_{R'_i} b$ (i.e., they share that $a \ge b$) then $a \ge_{f(R_i)} b \Leftrightarrow a \ge_{f(R'_i)} b$ (i.e., the choice is irrelevant). 

E.g.: $f(R) = R_1$ is IIA, since:
* if $i = 1$, then the property is trivially satisfied ($P \Rightarrow P$)
* if $i \ne 1$, then the choice of another person does not influence in any way $f(R)$, hence, also for irrelevant alternatives.

````python
import pandas as pd
import numpy as np

df_nodes = pd.read_csv('candidates.txt', delimiter=",")
profile = np.loadtxt('rankings.txt', delimiter=",", dtype='int16')

print(df_nodes)
mapping = {} 
for row in df_nodes.iloc:
    mapping[row['Label']] = row['RealName']
print(mapping)
print(profile)
````

````python
#@title Solution (plurality) { form-width: "10px" }
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import numpy as np


def plurality(profile):
    num_candidates = np.shape(profile)[1]
    votes = np.zeros(num_candidates)

    for row in profile:
        votes[row[0]] += 1

    rank = np.argsort(votes)[::-1]
    return rank[0], rank

elected, ranking = plurality(profile)
print(elected)
print(ranking)
print([mapping[index] for index in ranking])
````

````python
#@title Solution (approval-3) { form-width: "10px" }
def approval3(profile):
    num_candidates = np.shape(profile)[1]
    votes = np.zeros(num_candidates)

    for row in profile:
        for j in range(3):
            votes[row[j]] += 1

    rank = np.argsort(votes)[::-1]
    return rank[0], rank

elected, ranking = approval3(profile)


print(elected)
print(ranking)
print([mapping[index] for index in ranking])
````

````python
#@title Solution (veto) { form-width: "10px" }
def veto(profile):
    num_candidates = np.shape(profile)[1]
    votes = np.zeros(num_candidates)

    for row in profile:
        votes[row[-1]] += 1

    rank = np.argsort(votes)
    print(votes)
    return rank[0], rank

elected, ranking = veto(profile)

print(elected)
print(ranking)
print([mapping[index] for index in ranking])
````

````python
#@title Solution (copeland) { form-width: "10px" }
def copeland(profile):
    num_candidates = np.shape(profile)[1]

    pairwise = np.zeros(num_candidates)
    comparison = np.zeros((num_candidates, num_candidates))
    for p in profile:
        # Rank list of a voter p in profile
        p = list(p)

        # Fix the first candidate
        for i in range(0, num_candidates):
            c1 = p[i]
            for j in range(i + 1, num_candidates):
                c2 = p[j]
                # Check the winner between the two candidates for voter p
                if p.index(c1) < p.index(c2):
                    pairwise[c1] += 1
                    comparison[c1, c2] += 1
                else:
                    pairwise[c2] += 1
                    comparison[c2, c1] += 1

    drawGraph(pairwise, comparison, num_candidates)

    rank = np.argsort(pairwise)[::-1]
    return rank[0], rank
````

Build and draw a graph which represent the problem using  majority graph, implementing it in `networkx`.

````python
#@title Solution { form-width: "10px" }
def drawGraph(pair, comp, n, mapping=mapping):
    g = nx.DiGraph()

    for i in range(0, n):
        g.add_node(i)

    edges = []
    weights = []
    for i in range(0, n):
        for j in range(i+1, n):
            if comp[i, j] > comp[j, i]:
                edges.append((i, j))
                weights.append(comp[i, j]-comp[j, i])
            else:
                edges.append((j, i))
                weights.append(comp[j, i]-comp[i, j])
    g.add_edges_from(edges)

    plt.figure(figsize=(10, 10))
    options = {
        'width': 1,
        'with_labels': True,
        'font_weight': 'regular',
    }
    print(g.nodes())
    h = nx.relabel_nodes(g, mapping)
    print(h.nodes())
    pos  = nx.spring_layout(h, k=0.25, iterations=50)
    nx.draw(h, node_size=pair*10, pos=pos, **options)
    nx.draw_networkx_edge_labels(h, pos=pos, 
                                 edge_labels={(mapping[edge_pair[0]], mapping[edge_pair[1]]): int(weight) for edge_pair, weight in zip(edges, weights)})
    ax = plt.gca()
    ax.collections[0].set_edgecolor("#ffffff")
    plt.show()


voting_rule = 'copeland'
elected, ranking = copeland(profile)
print('The elected candidate is:')
print(df_nodes['RealName'][elected])


print(elected)
print(ranking)
print([mapping[index] for index in ranking])
````
